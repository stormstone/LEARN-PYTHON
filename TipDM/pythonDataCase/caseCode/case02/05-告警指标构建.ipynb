```python
import pandas as pd
import numpy as np
```

```python
alarm = pd.read_csv("E:/P-Python/alarm.csv",encoding="gb18030")
user = pd.read_csv("E:/P-Python/user.csv")
```

```python
alarm_num = alarm.groupby(['ID','date']).count()
```

```python
alarm_num.reset_index(inplace = True) #重新按行设置索引
```

```python
alarm_final_new = pd.merge(alarm_num,user,on = ["ID","date"],how = "outer")  #合并表,merge 根据一个或多个Key，将不同的DataFrame链接
```
